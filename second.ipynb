rename columns

```python
df=pd.DataFrame({'A col':[1,2,3,6,7,8,9,22], "B col":[4,5,6,7,8,9,10,11]})
df
```

```python
df.rename(columns={'A col':'col_a', 'B col':'col_b'}, inplace=True)
df
```

```python
df.columns=["columnaa", "columnbb"]
df
```

```python
#to replace character 
df.columns=df.columns.str.replace('_'," ")
df
```
