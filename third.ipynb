splitting a string into multiple columns

```python
import pandas as pd
df=pd.DataFrame({'name':['ali hussain','ahmed hassan','hassan raza','amna fahad'], 'location':['lahore, pakistan','islamabad, pakistan','pindi, afghanistan','karachi, dubai']})
df
```

```python
df[['firtsname','lastname']] = df.name.str.split(" ", expand=True)
df
```

```python
df[["city",'country']] = df.location.str.split(", ", expand=True)
df
```

```python
df=df[["firtsname",'lastname','city','country']]
df
```
